```python
print("Python Assignment - 1")
print("Himanshu Kadyan \n SAP ID - 500124212 \n BATCH - 51")

# Decimal to Binary
def decimal_to_binary(decimal):
    binary = ""
    while decimal > 0:
        remainder = decimal % 2
        binary = str(remainder) + binary
        decimal = decimal // 2
    return binary if binary else "0"

# Decimal to Octal
def decimal_to_octal(decimal):
    octal = ""
    while decimal > 0:
        remainder = decimal % 8
        octal = str(remainder) + octal
        decimal = decimal // 8
    return octal if octal else "0"

# Decimal to Hexadecimal
def decimal_to_hexadecimal(decimal):
    hexadecimal = ""
    hex_map = "0123456789ABCDEF"
    while decimal > 0:
        remainder = decimal % 16
        hexadecimal = hex_map[remainder] + hexadecimal
        decimal = decimal // 16
    return hexadecimal if hexadecimal else "0"

# Binary to Decimal
def binary_to_decimal(binary):
    decimal = 0
    power = 0
    for digit in reversed(binary):
        decimal += int(digit) * (2 ** power)
        power += 1
    return decimal

# Octal to Decimal
def octal_to_decimal(octal):
    decimal = 0
    power = 0
    for digit in reversed(octal):
        decimal += int(digit) * (8 ** power)
        power += 1
    return decimal

# Hexadecimal to Decimal
def hexadecimal_to_decimal(hexadecimal):
    decimal = 0
    power = 0
    hex_map = "0123456789ABCDEF"
    for digit in reversed(hexadecimal):
        decimal += hex_map.index(digit.upper()) * (16 ** power)
        power += 1
    return decimal

if __name__ == "__main__":
    print("Number System Converter")
    print("1. Decimal to Binary")
    print("2. Decimal to Octal")
    print("3. Decimal to Hexadecimal")
    print("4. Binary to Decimal")
    print("5. Octal to Decimal")
    print("6. Hexadecimal to Decimal")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        decimal = int(input("Enter decimal number: "))
        print("Binary:", decimal_to_binary(decimal))
    elif choice == 2:
        decimal = int(input("Enter decimal number: "))
        print("Octal:", decimal_to_octal(decimal))
    elif choice == 3:
        decimal = int(input("Enter decimal number: "))
        print("Hexadecimal:", decimal_to_hexadecimal(decimal))
    elif choice == 4:
        binary = input("Enter binary number: ")
        print("Decimal:", binary_to_decimal(binary))
    elif choice == 5:
        octal = input("Enter octal number: ")
        print("Decimal:", octal_to_decimal(octal))
    elif choice == 6:
        hexadecimal = input("Enter hexadecimal number: ")
        print("Decimal:", hexadecimal_to_decimal(hexadecimal))
    else:
        print("Invalid choice")

```
